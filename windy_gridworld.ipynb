# Initiate Agent and WindyGridWorld

```python
agent = Agent()
```

```python
agent.set_states(7,10)
agent.set_rewards((-1,-100,10))
agent.get_rewards()
agent.set_discount(1)
agent.set_full_actions()
agent.init_q_values()
```

## Initiate special grids

```python
agent.set_starting_grid((6,0))
agent.set_terminal_grid((3,7))
agent.set_death_grid((4,9))
```

## Initiate the columns with soft and strong wind

```python
for x in [3,4,5,8]:
    agent.set_soft_windy_columns(x)
    
for x in [6,7]:
    agent.set_hard_windy_columns(x)
    
print("WindyGridWorld Columns for the Soft Wind:", agent.get_soft_windy_cols())
```

```python
print("WindyGridWorld Columns for the Hard Wind:", agent.get_hard_windy_cols())
```

Here it is interesting to observe that SARSA did not find the optimal path and ends up in the death square. I believe that because the agent has a negative reward of -1 for every move and that it on average the reward from finishing in the terminal grid is less than going straight for the death grid, that the agent favor going to the death grid.

Because of the wind and the death square, it is impossible to reach the terminal state with the restricted actions. If we go from the left and try to reach from the bottom by going east, the agent ends up above the terminal square. The agent then needs to go to the top right of the grid then go down and hopefully go west to the terminal square but if the agent doesn't go south enough, the agent will get picked up by the wind and go back to the top of the grid. Therefore, the agent needs to go more south but there's now a death square and it dies. \

The agent cannot get to the terminal square with restricted amount of actions.
